# Importing the Dependencies

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# Data Collection & Processing

```python
credit_card_data = pd.read_csv('creditcard.csv')
```

# To get the first 5 Rows of Data

```python
credit_card_data.head()
```

# Dataset Information

```python
credit_card_data.info()
```

# Checking the Number of Missing Values in Dataset

```python
credit_card_data.isnull().sum()
```

# Distribution of Legit Transactions & Fraudulent Transactions

```python
credit_card_data['Class'].value_counts()
```

**This Data is highly unbalanced**

**0 --> Normal Transactions**

**1 --> Fradulent Transactions**

# Separating the Data for Analysis

```python
legit = credit_card_data[credit_card_data.Class == 0]
Fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
legit.shape
```

```python
Fraud.shape
```

# Statistical Measure about data

```python
legit['Amount'].describe()
```

```python
Fraud['Amount'].describe()
```

# Compare the values for both Transactions

```python
credit_card_data.groupby('Class').mean()
```

# Under-Samling


* Building a sample dataset containing similar distribution of Normal and Fradulent transactions

* No. of Fradulent transactions = 492

```python
legit_sample = legit.sample(n = 492)
```

# Concatenating 2 Dataframes

```python
new_dataset = pd.concat([legit_sample,Fraud], axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

# Splitting the Data into Features and Target

```python
X = new_dataset.drop('Class', axis = 1)
Y = new_dataset['Class']
```

# Splitting the data into Training and Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify = Y, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Model Training

```python
model = LogisticRegression()
```

# Training the Logictic regression Model with Training Data

```python
model.fit(X_train, Y_train)
```

# Model Evaluation on Accuracy Score

* Accuracy on training data

```python
X_train_prediction = model.predict(X_train)
```

```python
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy Score on training Data :",training_data_accuracy)
```

* Accuracy on test data

```python
X_test_prediction = model.predict(X_test)
```

```python
testing_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy Score on testing Data :",testing_data_accuracy)
```
